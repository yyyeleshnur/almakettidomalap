# Начальный анализ  

```python
import pandas as pd
import os

# Получение текущего рабочего каталога
current_directory = os.getcwd()
file_name = 'issues.csv'
file_path = os.path.join(current_directory, file_name)
data = pd.read_csv(file_path)

# Просмотр первых строк данных
print(data.head())

# Проверка типов данных и наличие пропущенных значений
print(data.info())
```

## Анализируем 

```python
# Исследование данных по темам (topic)
topic_counts = data['topic'].value_counts()
print("Распределение стихийных бедствий и проблем по темам:")
print(topic_counts)

# Анализ данных по странам (country)
country_counts = data['country'].value_counts()
print("Страны с наибольшим количеством стихийных бедствий и проблем:")
print(country_counts)

# Изучение тенденций по годам (year)
trends_by_year = data.groupby('year')['index_trends'].value_counts()
print("Тенденции стихийных бедствий и проблем по годам:")
print(trends_by_year)

# Визуализация данных 
import matplotlib.pyplot as plt

# График распределения стихийных бедствий по темам
plt.figure(figsize=(8, 6))
topic_counts.plot(kind='bar')
plt.title('Распределение стихийных бедствий по темам')
plt.xlabel('Тема')
plt.ylabel('Количество')
plt.show()
```

## Углубленный анализ 
### Исследование данных по темам (topic)

```python
import matplotlib.pyplot as plt

# Круговая диаграмма распределения стихийных бедствий по темам
plt.figure(figsize=(8, 8))
data['topic'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Распределение стихийных бедствий по темам')
plt.ylabel('')
plt.show()
```

## Анализ данных по странам (country)
### выясним, какие страны наиболее часто встречаются в данных и какие темы бедствий и проблем преобладают в этих странах.

```python
# Страны с наибольшим количеством стихийных бедствий и проблем
top_countries = data['country'].value_counts().head(10)
print("Топ стран по количеству стихийных бедствий и проблем:")
print(top_countries)

# Анализ тем по топовым странам
for country in top_countries.index:
    print(f"\nТемы стихийных бедствий и проблем в {country}:")
    print(data[data['country'] == country]['topic'].value_counts())
```

## Изучение тенденций по годам (year) и индексам (index_trends)
### Давайте определим, как менялись стихийные бедствия и экологические проблемы со временем.

```python
# Тенденции стихийных бедствий и проблем по годам
trends_by_year = data.groupby('year')['index_trends'].value_counts()
print("Тенденции стихийных бедствий и проблем по годам:")
print(trends_by_year)

# Визуализация тенденций по годам
plt.figure(figsize=(10, 6))
data.groupby('year')['index_trends'].value_counts().unstack().plot(kind='line', marker='o')
plt.title('Тенденции стихийных бедствий и проблем по годам')
plt.xlabel('Год')
plt.ylabel('Количество')
plt.legend(title='Тенденции', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
